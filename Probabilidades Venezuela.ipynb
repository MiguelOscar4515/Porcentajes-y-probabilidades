## Explicación <br>
* Es incorrecto buscar la probabilidad de que se encuentren cierta cantidad de nueves o que sea tan exacto, pues al hacer esto estamos considerando como espacio muestral a todos los números reales de `0` a `10058774`, cuando la cantidad de votos solo puede ser un número entero.
* Solo tomo como referencia a los dos votantes principales, pues incorrecto decir `es extraño que los 3 se redondeen a una cifra decimal`, esto porque si `a` y `b` se redondean cerca de de la primera cifra decimal, entonces, por consiguiente, `c = 10058774 - a - b` también lo va a estar. En términos probabilísticos `c` no es un evento independiente de los otros `2`.
* Para calcular la probalidad sería: $P(A \cap B) = P(A)P(B)$ porque `A` y `B` son eventos independientes.
* $P(A):$
  * Eventos favorables: 540
  * Eventos posibles: 6000001
* $P(B):$
  * Eventos favorables: 539
  * Eventos posibles: 6000000
* $P(A \cap B) = \frac{540}{6000001} \frac{539}{6000000} = 8.084998652500224e-09$
* Es súper pequeña pero al final es la misma para cualesquiera dos valores del intervalo tomados al azar.

**Esta función es para extraer los decimales del número, pues nos interesa saber cuáles son los números que tienen la misma propiedad que los porcientos dados**

```python
def extraer_decimales(numero):
    # Convertir el número a cadena
    numero_str = str(numero)
    # Dividir la cadena en la parte entera y la parte decimal
    partes = numero_str.split('.')
    # Verificar si hay una parte decimal
    if len(partes) == 2:
        return partes[1]
    else:
        return '0'
```

* `q`: Contiene todas las posibles terminaciones de los porcentajes (Solo tomo 5 porque el porcentaje alcanzado por Maduro solo tiene 4 `nueves` después del 1)

* En el `if` llamo a la función `extraer_decimales` y solo me quedo con los 5 primeros elementos de la cadena.

* Imprimo `i` que representa la cantidad de votos, así como su porcentaje del total

```python
count = 0
q = ['09999', '19999', '29999', '39999', '49999', '59999', '69999', '7999', '89999', '99999'] 
for i in range(2000000, 8000000):
    if extraer_decimales(100*i/10058774)[:5] in q:  
        print(i, 100*i/10058774)
        count += 1
print(count)
```
